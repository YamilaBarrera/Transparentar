```python
import pandas
from datetime import datetime, timedelta

#data = pandas.read_csv("partes-extendido.csv")
#data = pandas.read_csv("partes.csv")
data = pandas.read_csv("mat2mart.csv", delimiter=";")
```

```python
prop = "riesgo"
dataPlot = data.pivot_table(index='latitud', columns='longitud', values=prop, fill_value=0)
```

```python
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import matplotlib
import matplotlib.colors
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace, meshgrid

matplotlib.rcParams['figure.figsize'] = (12,12)
 
map = Basemap(projection='merc',
              resolution='l',
              llcrnrlon=-70, llcrnrlat=-57,
              urcrnrlon=-50, urcrnrlat=-33)

lats = dataPlot.index
lons = dataPlot.columns
map.drawcoastlines()
map.drawcountries()
map.drawmapboundary()
map.fillcontinents(color='lightgrey')

x, y = meshgrid(lons, lats)

cmap="autumn_r"
map.pcolormesh(x, y, dataPlot, latlon=True, cmap=cmap, vmin=0.000001)
plt.colorbar()

plt.show()
np.nanmax(dataPlot)
```
